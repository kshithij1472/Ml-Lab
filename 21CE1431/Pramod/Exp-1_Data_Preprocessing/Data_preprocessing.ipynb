#### Import library

```python
import os
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

##### CONSTANTS

```python
BASE_PATH = 'Dataset'
```

```python
sites = os.listdir(BASE_PATH)
sites, len(sites)
```

```python
data = None
for site in sites:
    data = pd.concat((data, pd.read_csv(os.path.join(BASE_PATH, site), skiprows=10)))
```

```python
data
```

```python
data.info()
```

```python
data.describe()
```

```python
y = data.pop('GHI (Wh/m2)')
data['GHI (Wh/m2)'] = y
data.columns
```

### Missing value

```python
data.columns
```

```python
# check missing values
for col in data.columns:
    if data[col].isnull().sum() > 0:
        print(col, ':', data[col].isnull().sum())
```

```python
data.drop(['Wind Speed at 3m (std dev) Uncertainty (m/s)'], axis=1, inplace=True)
data.columns
```

```python
# check missing values
for col in data.columns[4:]:
    val = data[col].mean()
    data[col].fillna(value=val, inplace=True)
```

#### Normalization

```python
num_cols = data.columns[4:-1]
num_cols
```

```python
data_proc = pd.DataFrame()
for col in num_cols:
    scaler = StandardScaler(with_mean=True, with_std=True)
    data_proc[col] = scaler.fit_transform(data[col].values.reshape(-1, 1)).reshape(-1)
    
```

```python
scaler = MinMaxScaler(feature_range=(-1, 1))
data_proc[data.columns[-1]] = scaler.fit_transform(data[data.columns[-1]].values.reshape(-1, 1)).reshape(-1)
    
```

```python
data_proc
```

```python
def get_class(ghi):
    if ghi > 0.4:
        return 'Running'
    elif ghi >= -0.4:
        return 'Monitoring'
    elif ghi >= -1:
        return 'Inspecting'
```

```python
data_proc['Class'] = data_proc[data.columns[-1]].apply(get_class)
data_proc
```

```python
data_proc.to_csv('processed_data.csv', index=False)
```
